So, und jetzt am allerbesten: Den Code schreibe ich in in einen String. Da kann ich dann zB auch Variablen aus dem Notebook einfügen (per f-Strings). Den String schicke ich dann (base64-codiert) an die URL und binde das Ergebnis in ein IFrame ein. Und zwar so:

````python
name = "Michael"

code = f"""
col = "lightblue"
n = 8

def setup():
    global col, n
    createCanvas(200, 200)
    background(190)
    fill(col)
    r = width/n
    for i in range(n):
        circle(i*r+r/2, 100, r)
    name = "{name}"
    textAlign(CENTER)
    textSize(20)
    fill(0)
    text(name, 100, 170)
""".strip()

from IPython.display import IFrame, display
import base64

def encode_code_to_url(code, base_url="") -> str:
    code_b64 = base64.b64encode(code.encode('utf-8')).decode('utf-8')
    return f"{base_url}{code_b64}"

def pyp5js_iframe(code, base_url, width=400, height=400) -> IFrame:
    """
    Generates an iframe for pyp5js with the given code.
    
    Parameters:
    - code: The p5.js code to be executed.
    - width: Width of the iframe.
    - height: Height of the iframe.
    - styleframe: Additional CSS styles for the iframe.
    
    Returns:
    An IFrame object to be displayed in a Jupyter notebook.
    """
    
    total_url = encode_code_to_url(code, base_url)
    frame = IFrame(src=total_url, width=width, height=height)
    # print(frame.src)
    return frame

url = f"https://berinhard.github.io/pyp5js/pyodide/?sketch="
display(pyp5js_iframe(code, base_url=url, width=800, height=400))
````

````python
fillcolor = "'pink'"  
example_text = "Hello, pyp5js!"

code = f"""
def setup():
    size(400, 400)

def draw():
    background(200)
    diameter = sin(frame_count / 60) * 50 + width / 2
    fill({fillcolor})
    circle(width / 2, height / 2, diameter)
    text_align(CENTER)     # py5 uses snake_case (unlike p5.js)
    text_size(20)
    fill(0)
    text("{example_text}", width / 2, height - 30)
""".strip()

fullscreen_url = f"https://abav.lugaralgum.com/pyp5js/py5mode/fullscreen.html?sketch="

display(pyp5js_iframe(code, fullscreen_url, width=800, height=450))
````

````python
from IPython.display import HTML
from myst_nb import glue

def make_btn(url: str, text: str="Im Online-Editor öffnen", kind: str="primary", size: str="sm"):
    return HTML(
        f'<a class="sd-btn sd-btn-{kind} sd-btn-{size}" '
        f'href="{url}" target="_blank" rel="noopener">{text}</a>'
    )

# url ist bei dir bereits definiert
glue("btn1", make_btn(url), display=False)  # kein Direkt-Output, nur „glue“
````
