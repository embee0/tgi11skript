## Lokaler p5js-Sketch

Achtung: Änderungen an Sketchen werden nicht automatisch übernommen, wenn die Seite neu geladen wird. Das liegt daran, dass der Browser die Datei im Cache speichert. Um das zu umgehen, hab ich bisher keine andere Möglichkeit gefunden als `make clean` und dann die lokale Seite mit geleertem Cache nue zu laden, also: Cmd+Shift+R (Mac) oder Strg+Shift+R (Windows/Linux).

Im Extremfall auch `make kill; make clean; make develop &` mit einem neuen lokalen Server.
