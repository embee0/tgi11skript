## Tooltips mit *tippy*

```{hint}
Damit die Tooltips funktionieren, muss `teachbooks_sphinx_tippy` in der Konfigurationsdatei aktiviert sein.

Damit auch die speziellen CSS-Klassen funktionieren, muss in der `_config.yml` der Key `tippy_add_class` auf `has-tippy` gesetzt werden. Und für diese Klasse wird der Datei `tippy_custom.css` ein paar Stile hinzugefügt, damit die Tooltips etwas anders aussehen, insb. die Glossareinträge erhalten dann eine Lupe als Icon.
```
