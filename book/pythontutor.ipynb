## Aus dem Jupyter-Notebook heraus aufrufen


````python
from IPython.display import IFrame
from urllib.parse import quote_plus

def python_tutor(code_str: str, height: int = 400, cur_instr: int = 0, code_div_height=400, code_div_width=400) -> IFrame:
    """Return an IFrame of the Python Tutor visualization of the given code."""
    code_str = code_str.strip()
    code_str = quote_plus(code_str)
    src = f"https://pythontutor.com/iframe-embed.html#code={code_str}"
    src += f"&curInstr={cur_instr}&codeDivHeight={code_div_height}&codeDivWidth={code_div_width}&cumulative=false&heapPrimitives=nevernest&origin=opt-frontend.js&py=3&rawInputLstJSON=%5B%5D&textReferences=false"
    # print(f"{src}")
    return IFrame(src=src, width="100%", height=height, frameborder=0)

code_bsp = """
original = [1, 2, 3]
liste2 = original.copy()
liste3 = original
liste3[0] = 42
# Was wird hier ausgegeben - und warum?
print(original[0])
"""

python_tutor(code_bsp, height=350, cur_instr=2, code_div_height=600, code_div_width=400)
````
