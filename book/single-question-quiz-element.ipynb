# Quizfragen mit single-question-quiz-element

```python
from IPython.display import HTML, display
import html

def render_quiz_markup(question, options, explanation=None, name="quiz-1"):
    opts = []
    for i, (text, correct, fb) in enumerate(options, 1):
        opts.append(f'''
        <div class="form-check">
          <input class="form-check-input" type="radio" name="quizOptions" id="opt-{name}-{i}" value="{i}" data-correct={"true" if correct else "false"} />
          <label class="form-check-label" for="opt-{name}-{i}">{html.escape(text)}</label>
          <aside class="alert {'alert-success' if correct else 'alert-danger'} mt-2" style="display:none">{html.escape(fb)}</aside>
        </div>
        ''')

    explanation_block = (
        f'<aside class="alert alert-secondary mt-3" style="display:none">{html.escape(explanation)}</aside>'
        if explanation else ""
    )

    display(HTML(f"""
<single-question-quiz-element name="{html.escape(name)}">
  <form class="bg-light rounded p-4">
    <div class="quiz-question">
      <p class="mb-4 fw-bold">{html.escape(question)}</p>
      {''.join(opts)}
      <div class="mt-3">
        <button type="submit" class="btn btn-primary" data-action="submit">Antwort einreichen</button>
        <button type="button" class="btn btn-secondary" style="display:none" data-action="reset">Neu starten</button>
      </div>
      {explanation_block}
    </div>
  </form>
</single-question-quiz-element>
"""))

# Beispiel
render_quiz_markup(
    "Welche Aussage trifft zu?",
    [
        ("Listen sind unveränderlich.", False, "Nope – das sind Tuples."),
        ("Tuples sind unveränderlich.", True, "Richtig!"),
        ("dicts waren schon immer geordnet.", False, "Verlässlich erst ab 3.7/3.8."),
    ],
    explanation="Kurzmerker: list (mutable), tuple (immutable), dict (insertion-ordered).",
    name="py-basics-q1",
)
```
